## Articles Embeddings

```python
import pandas as pd
```

```python
#read data folder
import os
dir_path = r'2020'
res = []

for file_path in os.listdir(dir_path):
    if os.path.isfile(os.path.join(dir_path, file_path)):
        res.append(os.path.join(dir_path, file_path))
```

```python
import xml.etree.ElementTree as ET
```

```python
#Read xml to df
data = []
for myxml in res:
    tree = ET.parse(myxml)
    root = tree.getroot()
    #data = []
    for elem in root:
        row = {}
        for subelem in elem:
            row[subelem.tag] = subelem.text
        data.append(row)
df = pd.DataFrame(data)
```
